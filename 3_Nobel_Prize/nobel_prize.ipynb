# History of Nobe Prize Winners

The Nobel Prize has been among the most prestigious international awards since 1901. Each year, awards are bestowed in chemistry, literature, physics, physiology or medicine, economics, and peace. In addition to the honor, prestige, and substantial prize money, the recipient also gets a gold medal with an image of Alfred Nobel (1833 - 1896), who established the prize.

The Nobel Foundation has made a dataset available of all prize winners from the outset of the awards from 1901 to 2023. The dataset used in this project is from the Nobel Prize API and is available in the `nobel.csv` file in the `data` folder.

#### 2.) Which decade had the highest ratio of US-born Nobel Prize winners to total winners in all categories?

```python
nobel['year'].unique()
```

```python
nobel['decade'] = np.floor(nobel['year']/10)*10
nobel['decade'] = nobel['decade'].astype('int64')
```

```python
total_winners = nobel.groupby('decade')['birth_country'].count().reset_index().rename(columns={'birth_country':'total_winners'})
total_winners
```

```python
filtered_winners = nobel[nobel['birth_country']=='United States of America']
us_winners = filtered_winners.groupby('decade')['birth_country'].count().reset_index().rename(columns={'birth_country':'us_winners'})
us_winners
```

```python
decades_df = total_winners.merge(us_winners, on='decade')
decades_df['us_ratio'] = (decades_df['us_winners'] / decades_df['total_winners']) * 100

decades_df.sort_values(by='us_ratio', ascending=False)
```

```python
max_decade_usa = 2000

print('Most US born Nobel Prize winners were in', max_decade_usa)
```

#### 3.) Which decade and Nobel Prize category combination had the highest proportion of female laureates?

```python
nobel['sex'].value_counts(dropna=False) # There are missing values.
```

```python
female_winners = nobel[nobel['sex']=='Female']
```

```python
total_count = nobel.groupby(['decade','category'])['full_name'].count().reset_index().rename(columns={'full_name':'total'})
female_count = female_winners.groupby(['decade','category'])['full_name'].count().reset_index().rename(columns={'full_name':'female'})
```

```python
gener_agg = total_count.merge(female_count, on=['decade','category'])

gener_agg['female_ratio'] = (gener_agg['female'] / gener_agg['total']) * 100

gener_agg.sort_values(by='female_ratio', ascending=False).head()
```
